```python
#!/usr/bin/env python
"""

Masked wordcloud
================

Using a mask you can generate wordclouds in arbitrary shapes/
"""

from os import path 
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import os


from wordcloud import WordCloud, STOPWORDS


#get data directory 

dir_path = r"C:\Users\Pangloss\Desktop\Pinwheel_Review.txt"
dir_path2 = r"C:\Users\Pangloss\Pictures\BGR_New-Logo-Gold.png"
dir_path3 = r"C:\Users\Pangloss\Desktop\pinwheel_masked.png"

#Read the whole text

Pinwheel_txt = open(dir_path).read()


# read the mask image

Pinwheel_mask = np.array(Image.open(dir_path2))

#alice_mask = np.expand_dims(alice_mask, axis=0)

#Assigning stopwords
stopwords = set(STOPWORDS)

stopwords.add("said")


wc = WordCloud(background_color="black", max_words=2000, mask=Pinwheel_mask,
               stopwords=stopwords, contour_width=3, contour_color='steelblue')

# generate word cloud
wc.generate(Pinwheel_txt)

# store to file
wc.to_file(dir_path3)

# show
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.figure()
#plt.imshow(Pinwheel_mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis("off")
plt.show()
```
